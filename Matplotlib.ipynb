```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("titanic.csv")
data
```

```python
x = data["PassengerId"]
y = data["Fare"]
plt.plot(x,y)
plt.xlabel("PassengerId")
plt.ylabel("Salary")
plt.title("PassengerId vs Salary")
plt.show()
```

# BAR PLOT

```python
plt.bar(x,y)
plt.xlabel("PassangerId")
plt.ylabel("salary")
plt.title("PassengerId vs Salary")
plt.show()
```

# scatter plot

```python
plt.scatter(x,y)
plt.xlabel("PassengerId")
plt.ylabel("salary")
plt.title("PassengerId vs Salary")
plt.show()
```
